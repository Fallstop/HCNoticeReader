```python
import numpy
from icalendar import Calendar, Event
from datetime import datetime
from pytz import UTC # timezone
import re
```

```python
g = open('schooldays.ics','rb')
gcal = Calendar.from_ical(g.read())
```

```python
events = []
startDate = datetime(2023, 5, 1)
for component in gcal.walk():
    if component.name == "VEVENT" and component.get('DTEND'):
        summary = component.get('SUMMARY')
        isSchool = re.search(r"Day ?((?:10)|[1-9])",summary,re.IGNORECASE+re.MULTILINE)
        time: datetime = (component.get('DTEND').dt)
        if isSchool and int(time.strftime("%Y%m%d%H%M%S")) >= int(startDate.strftime("%Y%m%d%H%M%S")):
            events.append(component)
```

```python
events.sort(key=lambda x: int(x.get('DTEND').dt.strftime("%Y%m%d%H%M%S")))
```

```python
cal = Calendar()
cal.add('prodid', '-//My calendar product//mxm.dk//')
cal.add('version', '2.0')
```

```python
week_rotation = 0
day_rotation = -1

for schoolDay in events:
    schoolDayWeekDay = schoolDay.get('dtstart').dt.weekday()
    if day_rotation >= schoolDayWeekDay:
        week_rotation += 1
        if week_rotation == 2:
            week_rotation = 0
    day_rotation = schoolDayWeekDay

    event = Event()
    name = f"{week_rotation}-{day_rotation}"
    event.add('summary', name)
    event.add('dtstart', schoolDay.get('dtstart'))
    event.add('dtend', schoolDay.get('dtend'))
    event.add('dtstamp', schoolDay.get('dtstamp'))
    cal.add_component(event)

    print(f"{component.get('SUMMARY')} - {schoolDay.get('dtstart').dt} - {schoolDay.get('dtstart').dt.weekday()} - {name}")
    
```

```python
f = open('output.ics', 'wb')
f.write(cal.to_ical())
f.close()
```
